 # Clasificacion

 Dado un conjunto de atributos (x) y un modelo de clasificacion -> etiqueas (y)

 ## Ejemplos

 - Evaluacion de riesgo
 - Detección de SPAM 
 - Detección de sentimiento
 - Identifiacion de tumores

## Componentes principales

- Conjunto de entrenamiento -> Input para entrenar
- Algoritmo de Clasificación -> Se obtiene -> Modelo de clasificacion
- Conjunto de validación -> Para probar el -> Modelo de clasificacion

## Funciona:

- Mejor con datos binarios o Nominaes
- Peor con datos ordinales (categorías ordenadas, ej: grande, mediano, pequeño)


## Clasificacion en ML vs DM

- En ML buscamos automatizar una tarea
- En DM buscamos encontrar un patrón en los datos

## Como saber si es un buen modelo

- Lo mas importante es la **capacidad predictiva**
- Pero predecir correctamente no es suficiente para determinar la capacidad predictiva
- Debe ser capaz de **generalizar**, ser capaz de realizar predicciones correctas en datos distintos al del entrenamiento

## Accuracy (Exactitud)

Accuracy = (TP + TN)/(TP+TN+FP+FN)

Limitaciónes: no es buena medida para clases desbalanceadas, ej: clase 1 tiene 9990 y clase 2 tiene 10, si el modelo clasifica todos como clase 1 tiene accuracy de 99.9

# Métodos de evaluacion de modelos

## Holdout

Más tipico, se particionan los datos: 2/3 para entenar, 1/3 para evaluar

Puede variar mucho dependiendo de la particion

## Random Subsampling

Se repite holdout varias veces

Puede que algunos datos nunca se usen para evaluar o entrenar

## Cross-validation




# Problemas prácticos en la clasificación

- Errores de entrenamiento, malos resultados sobre los datos de entrenamiento. Ocurre cuando el clasificador no tiene la capacidad de aprender el patrón
- Errores de generalizacióm, no puede predecir nuevos datos

## Overfitting

Se aprende demasiado los datos de entrenamiento, entonces no puede predecir datos nuevos

Es un reflejo de un modelo más complejo que lo necesario.

El error de entrenamiento no es un indicador confiable de cómo se desempeñaría el modelo sobre datos nuevos.

## Underfitting

Intenta generalizar demasiado, no se ajusta lo suficiente
