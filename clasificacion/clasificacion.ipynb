## Componentes principales

- Conjunto de entrenamiento -> Input para entrenar
- Algoritmo de Clasificacin -> Se obtiene -> Modelo de clasificacion
- Conjunto de validacin -> Para probar el -> Modelo de clasificacion

## Funciona:

- Mejor con datos binarios o Nominaes
- Peor con datos ordinales (categoras ordenadas, ej: grande, mediano, pequeo)


## Clasificacion en ML vs DM

- En ML buscamos automatizar una tarea
- En DM buscamos encontrar un patrn en los datos

## Como saber si es un buen modelo

- Lo mas importante es la **capacidad predictiva**
- Pero predecir correctamente no es suficiente para determinar la capacidad predictiva
- Debe ser capaz de **generalizar**, ser capaz de realizar predicciones correctas en datos distintos al del entrenamiento

## Accuracy (Exactitud)

## Holdout

Ms tipico, se particionan los datos: 2/3 para entenar, 1/3 para evaluar

Puede variar mucho dependiendo de la particion

## Random Subsampling

# Problemas prcticos en la clasificacin

- Errores de entrenamiento, malos resultados sobre los datos de entrenamiento. Ocurre cuando el clasificador no tiene la capacidad de aprender el patrn
- Errores de generalizacim, no puede predecir nuevos datos

## Overfitting

Se aprende demasiado los datos de entrenamiento, entonces no puede predecir datos nuevos

Es un reflejo de un modelo ms complejo que lo necesario.

El error de entrenamiento no es un indicador confiable de cmo se desempeara el modelo sobre datos nuevos.

## Underfitting

Intenta generalizar demasiado, no se ajusta lo suficiente
